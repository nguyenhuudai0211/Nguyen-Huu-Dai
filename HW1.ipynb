```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import numpy as np
url = "https://raw.githubusercontent.com/TungTh/tungth.github.io/master/data/vn-provinces.json"
try:
    map_data = gpd.read_file(url)
    col_name = 'name' if 'name' in map_data.columns else map_data.columns[1]
    print("Đã tải thành công bản đồ Việt Nam!")
except Exception as e:
    print("Mạng lỗi, không thể tải file GeoJSON. Đang tự động tạo bản đồ giả lập...")
    polygons = []
    names = []
    for i in range(4):
        for j in range(4):
            polygons.append(Polygon([(i, j), (i+1, j), (i+1, j+1), (i, j+1)]))
            names.append(f"Khu vực {i*4 + j + 1}")
    map_data = gpd.GeoDataFrame({'name': names, 'geometry': polygons})
    col_name = 'name'
tinh_list = map_data[col_name].dropna().unique()
df_business = pd.DataFrame({
    col_name: tinh_list,
    'Doanh_Thu': np.random.randint(100, 1000, size=len(tinh_list))
})
merged = map_data.merge(df_business, on=col_name, how="left")
fig, ax = plt.subplots(1, 1, figsize=(10, 12))
merged.plot(
    column='Doanh_Thu',
    ax=ax,
    legend=True,
    cmap='YlGnBu',
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': "Doanh thu (Tỷ đồng)", 'shrink': 0.5}
)

plt.title('23.4: BẢN ĐỒ CHOROPLETH PHÂN BỔ DOANH THU', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import folium
import pandas as pd
import numpy as np
from geopy.distance import geodesic

center_lat, center_lon = 10.7769, 106.7009
center_location = (center_lat, center_lon)

np.random.seed(42)
num_customers = 150

latitudes = center_lat + np.random.uniform(-0.12, 0.12, num_customers)
longitudes = center_lon + np.random.uniform(-0.12, 0.12, num_customers)

customers = pd.DataFrame({
    'CustomerID': range(1, num_customers + 1),
    'Lat': latitudes,
    'Lon': longitudes
})

def calculate_zone(row):
    dist_km = geodesic(center_location, (row['Lat'], row['Lon'])).kilometers
    if dist_km <= 3:
        return 'Vùng 1 (0-3km)', dist_km, 'green'
    elif dist_km <= 5:
        return 'Vùng 2 (3-5km)', dist_km, 'orange'
    elif dist_km <= 10:
        return 'Vùng 3 (5-10km)', dist_km, 'blue'
    else:
        return 'Ngoài vùng (>10km)', dist_km, 'red'

customers[['Zone', 'Distance_km', 'Color']] = customers.apply(
    lambda row: pd.Series(calculate_zone(row)), axis=1
)

m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

folium.Marker(
    location=[center_lat, center_lon],
    popup="<b>KHO HÀNG TRUNG TÂM</b>",
    icon=folium.Icon(color="darkred", icon="home", prefix='fa')
).add_to(m)

radii_config = {
    3000: {'color': 'green', 'name': '3 km'},
    5000: {'color': 'orange', 'name': '5 km'},
    10000: {'color': 'blue', 'name': '10 km'}
}

for radius_m, config in radii_config.items():
    folium.Circle(
        location=[center_lat, center_lon],
        radius=radius_m,
        color=config['color'],
        weight=2,
        fill=True,
        fill_opacity=0.1,
        popup=f"Vùng phục vụ: {config['name']}"
    ).add_to(m)
for _, row in customers.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Lon']],
        radius=4,
        color=row['Color'],
        fill=True,
        fill_opacity=0.7,
        popup=f"KH {row['CustomerID']}<br>Cách kho: {row['Distance_km']:.2f} km"
    ).add_to(m)
m
```

```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

place_name = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place_name, network_type='drive')

fig, ax = ox.plot_graph(G, node_size=1, node_color='w', edge_color='gray', edge_linewidth=0.5)

nodes, edges = ox.graph_to_gdfs(G)
print(f"Số lượng nút: {len(nodes)}")
print(f"Số lượng đường: {len(edges)}")

area = 7.72
network_density = len(edges) / area
print(f"Mật độ mạng lưới: {network_density:.2f} đoạn đường/km2")

avg_edge_length = edges['length'].mean()
print(f"Chiều dài trung bình đoạn đường: {avg_edge_length:.2f} mét")

print("\nVai trò của dữ liệu mạng lưới giao thông trong AI đô thị:")
print("1. Tối ưu hóa lộ trình: AI sử dụng đồ thị này để tìm đường ngắn nhất (Dijkstra/A*).")
print("2. Dự báo tắc nghẽn: Phân tích lưu lượng trên các đoạn đường để dự đoán điểm nóng giao thông.")
print("3. Quy hoạch đô thị: Xác định các 'nút thắt' cần mở rộng hoặc điều chỉnh tín hiệu đèn.")
print("4. Giao thông công cộng: Thiết kế vị trí trạm xe buýt tối ưu dựa trên mật độ nút giao.")
```

```python
import osmnx as ox
import networkx as nx
import random

place_name = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place_name, network_type='drive')

nodes = list(G.nodes())
start_node = random.choice(nodes)
end_node = random.choice(nodes)

path_dijkstra = nx.shortest_path(G, start_node, end_node, weight='length')
path_astar = nx.astar_path(G, start_node, end_node, weight='length')

print(f"Đường đi Dijkstra: {path_dijkstra}")
print(f"Đường đi A*: {path_astar}")

fig, ax = ox.plot_graph_routes(
    G,
    routes=[path_dijkstra, path_astar],
    route_colors=['red', 'blue'],
    route_linewidths=[6, 2],
    node_size=0
)

print(f"Độ dài Dijkstra: {nx.path_weight(G, path_dijkstra, weight='length'):.2f} mét")
print(f"Độ dài A*: {nx.path_weight(G, path_astar, weight='length'):.2f} mét")
print("\nPhân tích:")
print("- Dijkstra: Đảm bảo tìm đường ngắn nhất tuyệt đối bằng cách duyệt rộng.")
print("- A*: Tìm nhanh hơn nhờ sử dụng hàm heuristic để định hướng tìm kiếm.")
```

```python
import numpy as np
import folium
from sklearn.cluster import KMeans

# Dữ liệu giả lập
np.random.seed(42)
customer_data = np.random.uniform(low=10.75, high=10.80, size=(100, 2))

# Phân cụm
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(customer_data)
centers = kmeans.cluster_centers_

# Khởi tạo bản đồ
m = folium.Map(location=[10.7769, 106.7009], zoom_start=13)

# Thêm khách hàng
for point in customer_data:
    folium.CircleMarker(
        location=point,
        radius=3,
        color='blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

# Thêm kho đề xuất
for center in centers:
    folium.Marker(
        location=center,
        popup="Kho đề xuất",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)

# Xuất bản đồ
m.save("map.html")
m
```
